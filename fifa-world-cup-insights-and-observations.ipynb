```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/fifa-world-cup-data/WorldCupPlayers.csv")
```

```python
# Visualization 1: Histogram of Shirt Numbers
plt.figure(figsize=(10, 6))
plt.hist(df['Shirt Number'], bins=range(1, 24), color='skyblue')
plt.title('Distribution of Shirt Numbers')
plt.xlabel('Shirt Number')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Bar chart of Top 10 Coaches by Frequency
top_coaches = df['Coach Name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_coaches.plot(kind='bar', color='skyblue')
plt.title('Top 10 Coaches by Frequency')
plt.xlabel('Coach Name')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 3: Pie chart of Player Positions
plt.figure(figsize=(8, 8))
df['Position'].value_counts().plot(kind='pie', colors=['skyblue', 'lightblue', 'deepskyblue'], autopct='%1.2f%%')
plt.title('Distribution of Player Positions')
plt.ylabel('')
plt.show()
```

```python
# Visualization 4: Line plot of Goals Scored per Tournament
goals_per_tournament = df.groupby('RoundID')['Event'].apply(lambda x: x.str.count('G').sum()).reset_index()
plt.figure(figsize=(10, 6))
plt.plot(goals_per_tournament['RoundID'], goals_per_tournament['Event'], marker='o', color='skyblue')
plt.title('Goals Scored per Tournament')
plt.xlabel('Tournament')
plt.ylabel('Total Goals')
plt.grid(True)
plt.show()
```

```python
# Visualization 5: Box plot of Shirt Numbers
plt.figure(figsize=(10, 6))
plt.boxplot(df['Shirt Number'], patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box plot of Shirt Numbers')
plt.ylabel('Shirt Number')
plt.show()
```

```python
# Visualization 6: Bar chart of Top 10 Teams by Match Count
top_teams = df['Team Initials'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_teams.plot(kind='bar', color='skyblue')
plt.title('Top 10 Teams by Match Count')
plt.xlabel('Team Initials')
plt.ylabel('Match Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 7: Stacked bar chart of Events by Position
events_by_position = df.groupby('Position')['Event'].count().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(events_by_position['Position'], events_by_position['Event'], color='skyblue')
plt.title('Events by Position')
plt.xlabel('Position')
plt.ylabel('Event Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 8: Line plot of Goals Scored by Match
goals_by_match = df.groupby('MatchID')['Event'].apply(lambda x: x.str.count('G').sum()).reset_index()
plt.figure(figsize=(10, 6))
plt.plot(goals_by_match['MatchID'], goals_by_match['Event'], marker='o', color='skyblue')
plt.title('Goals Scored by Match')
plt.xlabel('Match')
plt.ylabel('Goals Scored')
plt.grid(True)
plt.show()
```

```python
# Visualization 9: Bar chart of Events by Coach
events_by_coach = df.groupby('Coach Name')['Event'].count().reset_index().sort_values(by='Event', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(events_by_coach['Coach Name'], events_by_coach['Event'], color='skyblue')
plt.title('Top 10 Coaches by Event Count')
plt.xlabel('Coach Name')
plt.ylabel('Event Count')
plt.xticks(rotation=90)
plt.show()
```
